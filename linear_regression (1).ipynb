```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/headbrain.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
x = df[["Head Size(cm^3)"]]
y = df[["Brain Weight(grams)"]]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,train_size= 0.8)
```

```python
from sklearn.linear_model import LinearRegression

ss = LinearRegression()
reg = ss.fit(x_train,y_train)

y_pred = reg.predict(x_test)

```

```python
x_test
```

```python
y_pred
```

```python
y_test
```

```python
import matplotlib.pyplot as plt

plt.scatter (x_test,y_pred,color = "cyan")
plt.scatter (x_test,y_test,color = "blue")
```

```python
from sklearn.metrics import r2_score

print(r2_score (y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error

print(mean_squared_error(y_test,y_pred))
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv("/content/cars.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
x = df[["Volume","Weight"]]
y = df[["CO2"]]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2)
```

```python
from sklearn.linear_model import LinearRegression
df = LinearRegression()
ss = df.fit(x_train,y_train)
y_pred = ss.predict(x_test)
```

```python

x_volume = x_test["Volume"]
x_weight = x_test["Weight"]

plt.scatter(x_volume, y_pred, color="yellow", label="Volume")
plt.scatter(x_weight, y_pred, color="blue", label="Weight")
plt.xlabel("Volume and Weight")
plt.ylabel("Predicted CO2")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score

print(r2_score(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
print(mse)
print(np.sqrt(mse))
```
